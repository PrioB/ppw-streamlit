```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat grafik kosong
G = nx.Graph()

# Menambahkan node dari 1 sampai 9
nodes = list(range(1, 10))
G.add_nodes_from(nodes)

# Menambahkan edges (hubungan) antar node
edges = [(1, 4), (1, 3), (1, 2), (2, 3), (3, 4), (4, 5), 
         (4, 6), (5, 6), (5, 8), (5, 7), (6, 7), (6, 8), (7, 8), (7, 9)]
G.add_edges_from(edges)

# Menggambar grafik
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Mengatur posisi node
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_color='black', font_weight='bold', edge_color='gray')
plt.title("Grafik Angka 1 sampai 9", fontsize=18)
plt.show()
```

```python
# Menghitung closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Menampilkan hasil closeness centrality
print("Closeness Centrality:")
for node, centrality in closeness_centrality.items():
    print(f"Node {node}: {centrality:.4f}")
```
